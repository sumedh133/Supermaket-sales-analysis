# Univariate Analysis of Dataset

```python
#categorical column and numerical columns seperation
cat=[]
num=[]
for col in data.columns:
    if data[col].nunique() > 10:
        num.append(col)
    else:
        cat.append(col)
```

On observing the dataset closely we find that there are only 3 cities and 3 branches respectively

```python
data['Branch'].value_counts()
```

```python
data['City'].value_counts()
data.head()
```

```python
# Branch A -> Yangaon city
# Branch B -> Mandalay city
# Branch C -> Naypyitaw city
# This revelation makes city column redundant, thus we will drop it out of the data
```

```python
mapp={'A':"A/Yangaon",'B':"B/Mandalay",'C':"C/Naypyitaw"}
data['Branch'].replace(mapp,inplace=True)
# data.rename(columns={"Branch":"Branch/City"},inplace=True)
data.drop('City',axis=1,inplace=True)
data.head()
```

# 3.Find distribution of ratings

```python
sns.histplot(data['Rating'],bins=35,kde=True)
```

```python
data['Rating'].skew()
```

```python
product_performance = data.groupby('Product line').agg({'Rating': 'mean', 'gross income': 'mean'})
product_performance = product_performance.sort_values('Rating')
```

```python
plt.figure(figsize=(10, 6))
plt.barh(product_performance.index, product_performance['Rating'], color='skyblue')
plt.xlabel('Average Rating')
plt.ylabel('Product Line')
plt.title('Average Rating by Product Line')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.barh(product_performance.index, product_performance['gross income'], color='lightcoral')
plt.xlabel('Average Gross Income')
plt.ylabel('Product Line')
plt.title('Average Gross Income by Product Line')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()
```

Here we can see that no product line is performing very poorly in comparision to the others

# Date analysis analysis

Income generated through time frame

```python
data_date=data.set_index("Date")
data_date.sort_index()
median_data = data_date.groupby(data_date.index).median()
median_data.head()
#median numerical data as per date
```

```python
median_data1=median_data['gross income']
median_data1.iplot(kind='line')
```

Lets see max income generated according to branches 

```python
max_data = data.groupby(['Date', 'Branch'])['gross income'].max().reset_index()
max_data.head()
```

```python
fig = px.line(max_data, x='Date', y='gross income', color='Branch', title='Max Gross Income by Branch over Time')
fig.update_layout(xaxis_title='Date', yaxis_title='Max Gross Income')
fig.show()
```

Cumulative quantity of goods sold on different dates

```python
quantity=data_date['Quantity']
def cumulative_data(group):
    return group.sum()
cumulative_quantity = quantity.groupby(data_date.index).apply(cumulative_data)
cumulative_quantity.head()
```

```python
cumulative_quantity.iplot(kind='line')
```

```python
cumulative_data_perBranch = data.groupby(['Date', 'Gender'])['Quantity'].sum().reset_index()
cumulative_data_perBranch.head()
```

```python
fig = px.line(cumulative_data_perBranch, x='Date', y='Quantity', color='Gender', title='Cumulative goods sold by Branch over Time')
fig.update_layout(xaxis_title='Date', yaxis_title='Goods sold')
fig.show()
```
